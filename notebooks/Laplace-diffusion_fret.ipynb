```python
import numpy as np
from scipy.stats import norm
import pylab
%matplotlib inline
```

```python
def evolve(u, d, fd, idr2, dt):
    """
    This evalutes the derivatives of the Laplacian
    of the Haas-Steinberg equation and calculates
    the excited state population `p` at a later time-step
    given the distibution of the previous time-step `pi`
    
    u   : distirbution at time-step
    d   : mutal diffusion coefficent
    idr2: inverse squared grid spacing
    dt  : time-step
    fd  : deactivation rate by FRET and S1->S0 transition (1.0 / tau0 *(1.0+(forster_radius/rda[1:-1])**6))
    """
    u[1:-1] += (d * (u[2:] - 2*u[1:-1] + u[0:-2]) - fd * u[1:-1]) * dt
    u[0] = 0
    u[-1] = 0


def get_initial_distibution(mean, sigma, n_points=100, verbose=False):
    """
    This initializes the probabilities of the vector `p` with the
    respective distances `r` given a the mean `rda_mean` and the
    width `sigma` of a normal-distribution
    """
    # linear
    n_points += 2
    a = float(0.5 * (mean * 2.0 - sigma / np.sqrt(1.0/12.0)))
    b = float(mean * 2.0 - a)
    if verbose:
        print "lower bound: %s" % a
        print "upper bound: %s" % b
        print "n points   : %s" % n_points
    rda = np.linspace(a, b, n_points)
    p = np.ones_like(rda)
    p /= sum(p)
    p[0], p[-1] = 0, 0
    return rda, p
    

def integrate_fret(tau0, rda_mean, sigma, rda_n_step, forster_radius, diffusion_coefficent,
                   simulation_time=5, verbose=False, dt=None):
    """
    Best use of all units in Angstrom and nano-seconds
    
    tau0 : lifetime of the dye without FRET
    rda_mea: mean donor-acceptor distance
    sigma: width of the rda distribution
    rda_min: minimum rda used for PDE integration
    rda_max: maximum rda used for PDE integration
    rda_n_step: the number of steps between min and max
    forster_radius: the forster radius of the dye pair
    diffusion_coefficient: the net diffusion coefficent of the dye-pair
    simulation_time: the simulation time 
    """
    rda, u = get_initial_distibution(rda_mean, sigma)
    
    d  = diffusion_coefficent
    fd = 1.0 / tau0 *(1.0+(forster_radius/rda[1:-1])**6)
    max_rate = max(d, max(fd))
    idr2 = 1. / (rda[1] - rda[0])
    if dt is None:
        dt = 1.0 / (2.0 *20* max_rate)
    
    time_steps = int(simulation_time / dt)

    if verbose:
        print "Diffusion coefficent [A2/ns]: %s" % d
        print "time-step: %s" % dt
        print "dr2: %s" % dr2
    
    # Set the initial conditions
    pt = np.zeros((time_steps, u.shape[0]))
    pt[0, :] = u
    for m in range(0, time_steps - 1):
        pt[m, :] = u
        evolve(u, d, fd, idr2, dt)
    time_axis = np.arange(0, time_steps) * dt
    return time_axis, rda, pt
        
```

```python
time_axis, rda, pt = integrate_fret(rda_n_step=100, 
                                    rda_mean=45.0, sigma=6.0,
                                    tau0=4.0, forster_radius=50, 
                                    diffusion_coefficent=10.,
                                    simulation_time=30, 
                                    verbose=False)

# Molecules that are in the S1-state
for i in range(0, time_axis.shape[0], 300):
    pylab.plot(rda, pt[i])
pylab.show()


# Molecules that are in the S1-state / normalized
for i in range(0, time_axis.shape[0], 300):
    pylab.plot(rda, pt[i] / sum(pt[i]))
pylab.show()

# Fraction of initial concentration
pylab.plot(time_axis, pt.sum(axis=1))
pylab.show()

# The mean position of the peak
times_of_calc_mean = list()
mean_distances = list()
for i in range(0, time_axis.shape[0], 300):
    p = pt[i, :]
    p /= sum(p)
    rda_mean = np.dot(rda, p)
    times_of_calc_mean.append(time_axis[i])
    mean_distances.append(rda_mean)

pylab.plot(times_of_calc_mean, mean_distances)
pylab.show()

# Expexred distribution of rates
p_expected = pt.sum(axis=0)
pylab.plot(rda, p_expected)
pylab.show()

print "mean distance: %s" % np.dot(rda, p_expected / sum(p_expected))
```
